```python
## 12

n = int(input("Enter a number to calculate factorial: "))
count = n
fact = 1

while count>=1:
    fact *= count
    count -= 1
print(f"Factorial of {n} = {fact}")
```

```python
## 13

num = int(input("Enter a number"))
temp = num
sum_of_didgits = 0

while num>0:
    rem = num%10
    sum_of_didgits += rem
    num //= 10
print(f"Sum of digits of {temp} = {sum_of_didgits}")
```

```python
## 14

num = int(input("Enter a number"))
is_prime = True
if num<=1:
    is_prime = False
else:
    for i in range(2, int(num**0.5)+1):
        if num%i==0:
            is_prime = False
            break
if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```
