List Comprehension

```python
## Without using list comprehension

lst = []
for x in range(10):
    lst.append(x**2)
print(lst)
```

```python
## With using list comprehension

squares = [x**2 for x in range(10)]
print(squares)
```

```python
## List comprehension with conditions

## normally

lst = []
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)
```

```python
## using list comprehension

evens = [x for x in range(10) if x%2==0]
print(evens)
```

```python
odds = [x for x in range(10) if x%2!=0]
print(odds)
```

```python
## Nested list comprehension

lst_1 = [1,2,3]
lst_2 = ['a', 'b', 'c']

pairs = [[i, j] for i in lst_1 for j in lst_2]
print(pairs)
```

```python
## List comperehension with function call

words = ["Hello","python","list","comprehension"]

lengths = [len(word) for word in words]

print(lengths)
```
