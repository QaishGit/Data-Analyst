## Sets

### Creating Empty Sets

```python
empty_set = set()
print(empty_set, type(empty_set))
```

### Creating Sets

```python
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_set = {1, 'hello', 3.14}
print(my_set)
```

```python
## From list to set 

my_set = set([1, 2, 3, 4, 5, 6])
print(my_set)
```

```python
## Eliminate duplicate elements

my_set = {1, 2, 3, 5, 6, 4, 5, 6}
print(my_set)
```

### Basic Set Operations

#### Addition

```python
my_set = set(range(1,7))
my_set.add(7)
print(my_set)
my_set.add(4)
print(my_set)
```

#### Removing an element

```python
my_set.remove(3)
print(my_set)
```

#### Removing an element that is not present in set

```python
my_set.remove(10)
print(my_set)
```

#### Removing an element from set that is not present without any error

```python
my_set.discard(11)
print(my_set)
```

#### pop method

```python
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

#### Clearing all the elements of set

```python
my_set.clear()
print(my_set)
```

### Set Membership test

```python
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)
print(10 in my_set)
```

### Mathematical Operations

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}
```

#### Union operation

```python
union_set = set1.union(set2)
print(set1)
print(set2)
print(f"Union of both sets: {union_set}")
```

#### Intersection operation

```python
intersection_set = set1.intersection(set2)
print(set1)
print(set2)
print(f"Intersection of both sets: {intersection_set}")
```

#### intersection_update

```python
set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}
```

#### Difference operation

```python
difference_set = set1.difference(set2)
print(set1)
print(set2)
print(f"Difference(set1 - set2): {difference_set}")
```

#### difference_update

```python
set1.difference_update(set2)
print(set1)
```

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}
```

#### Symmetric Difference 

```python
symm_diff_set = set1.symmetric_difference(set2)
print(set1)
print(set2)
print(f"Symmetric difference of both sets: {symm_diff_set}")
```

### Set Methods

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}
```

#### Checking subset

```python
print("Is set1 a subset of ste2: ", set1.issubset(set2))
print("Is set2 a subset of set1: ", set2.issubset(set1))
```

#### Checking superset

```python
print(f"Is set1 a superset of set2: {set1.issuperset(set2)}")
print(f"Is set2 a superset of set1: {set2.issuperset(set1)}")
```
