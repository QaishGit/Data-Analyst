```python
##Arithmetic Operators

a=10
b=5

add_result = a+b    ##addition
sub_result = a-b    ##subtraction
mult_result = a*b   ##multiplication
div_result = a/b    ##division
floor_result = a//b ##floor division operation
mod_result = a%b    ##modulus operation
expo_result = a**b  ##exponential operation

print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_result)
print(mod_result)
print(expo_result)
```

Comparision Operators

```python
##Equal to (==)
a=10
b=10
a==b

str1="Cash"
str2="Cash"
str1==str2
```

```python
##Not eqal to (!=)
str1="Cash"
str2="cash"
str1!=str2
```

```python
##Greater than (>)
a=45;b=55
print(a>b)

##Less than
print(a<b)
```

```python
##Greater than or equal to (>=)
a=45;b=45
print(a>=b)

##Less than or equal to (<=)
print(a<=b)
a=44
print(a<=b)
```

Logical Operators

```python
## AND, OR and NOT
##AND
x=True
y=True
print(x and y)
y=False
print(x and y)
```

```python
##OR
x=True
y=False
print(x or y)
x=False
print(x or y)
```

```python
##NOT
x=True
print(not x)
print(not(not x))
```

```python
##Simple Calculator

x = float(input("Enter first number: "))
y = float(input("Enter second number: "))

add_result = x+y
sub_result = x-y
mult_result = x*y
div_result = x/y
floor_div_result = x//y
mod_result = x%y
expo_result = x**y

print("Addition: ", add_result)
print("Subtraction: ", sub_result)
print("Multiplication: ", mult_result)
print("Division: ", div_result)
print("Floor Division: ", floor_div_result)
print("Modulus: ", mod_result)
print("Exponential: ", expo_result)
```
