```python
##While loop

count = 0

while count<5:
    print(count)
    count = count+1
```

```python
count = 0

while count%2==0:
    print(count)
    count = count+1
```

```python
##Loop Control Statements
##break statement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##Continue statement

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
##pass statement

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
##Nested Loops

for i in range(3):
    for j in range(2):
        print(f"i={i} and j={j}")
```

```python
##Examples

##Example 1 - Calculate sum of first n natural numbers using while loop

n = int(input("Enter a natural number to which you want sum: "))
count = 1
sum = 0

while count<=n:
    sum = sum + count
    count = count+1
print(f"Sum of first {n} natural numbers = {sum}")
```
