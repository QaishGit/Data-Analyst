```python
## 6

import random 

random_numbers = [random.randint(1,20) for _ in range(15)]
print(f"Original List: {random_numbers}")

sorted_numbers_asc = sorted(random_numbers)
print(f"Sorted list of numbers in ascending order: {sorted_numbers_asc}")

sorted_numbers_desc = sorted(random_numbers, reverse=True)
print(f"Sorted list of numbers in descending order: {sorted_numbers_desc}")

unique_numbers = list(set(random_numbers))
print(f"List of numbers without duplicates: {unique_numbers}")
```

```python
## 8

students = [
    {'name': 'Alice', 'score': 88},
    {'name': 'Bob', 'score': 72},
    {'name': 'Charlie', 'score': 95},
    {'name': 'David', 'score': 65},
    {'name': 'Eve', 'score': 78}
]

sorted_students_desc = sorted(students, key=lambda x: x["score"], reverse=True)
print("Sorted elements of students list in descending order using score: ")
for i in sorted_students_desc:
    print(i)
```

```python
## 9

def transpose_matrix(matrix):
    transposed = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    return transposed

matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

transposed = transpose_matrix(matrix)

print("Original matrix: ")
for row in matrix:
    print(row)

print("Transposed matrix: ")
for row in transposed:
    print(row)
```

```python
## 10

def flatten_matrix(nested_list):
    flattened = [item for sub_list in nested_list for item in sub_list]
    return flattened

nested_list = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

flattened = flatten_matrix(nested_list)

print(f"Original nested_list: {nested_list}")
print(f"Flattened nested list: {flattened}")
```

```python
## 13

import random

def reverse_list(numbers):
    return numbers[::-1]

numbers = [random.randint(1,100) for _ in range(10)]

reversed_numbers = reverse_list(numbers)

print(f"Original list: {numbers}")
print(f"Reversed list: {reversed_numbers}")
```

```python
## 14

def rotate_list(lst, n):
    return lst[n:] + lst[:n]

original_list = [1, 2, 3, 4, 5]
n = int(input("Enter the places by which you want to rotate the list(enter number less than 5): "))
rotated_list = rotate_list(original_list, n)
print(f"Original list: {original_list}")
print(f"Rotated list: {rotated_list}")
```

```python
## 15

import random

def intersect_list(lst1, lst2):
    #return [{i for i in lst1 for j in lst2 if i==j}]
    return [{i for i in lst1 if i in lst2}]

list_1 = [random.randint(1,50) for _ in range(10)]
list_2 = [random.randint(1,50) for _ in range(10)]

intersected_list = intersect_list(list_1, list_2)
print(f"First list: {list_1}")
print(f"Second list: {list_2}")
print(f"Intersected list: {intersected_list}")
```
