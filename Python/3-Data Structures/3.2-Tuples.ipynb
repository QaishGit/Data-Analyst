Tuples

Creating Tuples

```python
## Creating empty tuples

empty_tpl = ()

## OR

empty_tpl_1 = tuple()

print(empty_tpl, type(empty_tpl))
print(empty_tpl_1, type(empty_tpl_1))
```

```python
tpl = (1, 2, 3, 4, 5)
print(tpl, type(tpl))
```

```python
## Creating tuple from a list

numbers = tuple([1, 2, 3, 4, 5, 6])
print(numbers, type(numbers))
```

```python
mixed_tpl = (1, "Hello", 3.14, True)
print(mixed_tpl)
```

Accessing Tuple Elements

```python
numbers = (1, 2, 3, 4, 5, 6)
print(numbers[0])
print(numbers[len(numbers)//2])
print(numbers[-1])
```

Tuple Slicing

```python
numbers
```

```python
print(numbers[1:3])
print(numbers[::])
print(numbers[2:])
print(numbers[:4])
print(numbers[::-1])
print(numbers[::-2])
```

Tuple Operations

```python
## Concatination

concatinated_tpl = numbers + mixed_tpl
print(concatinated_tpl)
```

```python
print(mixed_tpl * 3)
```

```python
print(numbers * 3)
```

Immutable nature of tuples

```python
tpl = (1, 2, 3, 4, 5)
tpl[1] = "cash"
print(tpl)
```

Tuple Methods

```python
## Count method

numbers
print(numbers)
print("Count of element 2 in numbers: ", numbers.count(2))
```

```python
## Index method

print("Index of element 3 in numbers: ", numbers.index(3))
```

Tuple Packing and Unpacking

```python
## Packing

packed_tpl = 1, "Hello", 3.14
print(packed_tpl)
```

```python
## Unpacking

a, b, c = packed_tpl
print(a)
print(b)
print(c)
```

```python
## Unpacking with *

numbers = 1, 2, 3, 4, 5, 6
print(numbers)

first, *middle, last = numbers

print(first)
print(middle)
print(last)
```

Nested Tuple

```python
## Can create tuple inside a list

nested_lst = [[1, 2, 3], [4, 5, 6], (1, "Hello", 3.14)]
print(nested_lst[2])
print(nested_lst[2][:2])
```

```python
## Creating nested tuple

nested_tpl = ((1, 2, 3), ('a', 'b', 'c'), (True, False))

print(nested_tpl[0])
print(nested_tpl[1][2])
```

```python
## Iterating over a nested tuple

for sub_tpl in nested_tpl:
    for item in sub_tpl:
        print(item, end=" ")
    print()
```
