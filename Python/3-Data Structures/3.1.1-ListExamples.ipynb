### Example 1 - Create a To Do List to keep track of tasks

```python
## Creating a To Do List with some tasks that needed to be complete
to_do_list = ["Buy Groceries", "Clean the home", "Pay Bills"]

## Adding tasks to Tpo Do List 
to_do_list.append("Schedule Meeting")
to_do_list.append("Go for a run")

## Removing completed tasks
to_do_list.remove("Clean the home")

## Checking if a task is in the list
if "Pay Bills" in to_do_list:
    print("Don't forget to pay the utility bills")

## Printing remaining tasks
print("To Do List remaining tasks")
for task in to_do_list:
    print(f"-{task}")
```

### Example 2 - Create a list to store and calculate average, highest and lowest grades of a student

```python
## Organizing student grades
grades = [85, 92, 78, 90, 88]

## Adding a new grade
grades.append(95)

## Printing all the grades 
print("All the grades are:")
for grade in grades:
    print(grade, end=" ")
print()
## Calculating average grades
average_grade = sum(grades)/len(grades)
print(f"Average of grades = {average_grade}")

## Calculating highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest grade = {highest_grade}")
print(f"Lowest grade = {lowest_grade}")
```

### Example 3 - Use a list to collect and analyze user feedbacks

```python
## Collecting User feedback
feedbacks = ["Great Service!", "Very Satisfied", "Could be better", "Excellent experience"]

## Adding new feedback
feedbacks.append("Not happy with the service")

## Counting specific feedbacks
positive_feedbacks_count = sum(1 for feedback in feedbacks if "great" in feedback.lower() or \
                               "excellent" in feedback.lower())
print(f"Number of positive feedbacks = {positive_feedbacks_count}")

## Printing all feedbacks
print("All feedbacks")
for feedback in feedbacks:
    print(f"-{feedback}")
```
