## Dictionary

### Creating empty dictionaries

```python
empty_dict = {}
print(empty_dict)
print(type(empty_dict))

## OR

empty_dict_1 = dict()
print(empty_dict_1)
print(type(empty_dict_1))
```

### Creating dictionary with elements

```python
students = {"Name": "cash", "age": 25, "grade": 'A'}
print(students)
```

#### Only unique elements as key


```python
students = {"Name": "cash", "age": 25, "Name": 'A'}
print(students)
```

### Accessing dictionary elements

```python
## Using normal method

students = {"Name": "cash", "age": 25, "grade": 'A'}
print(students["grade"])
print(students["age"])
```

```python
## Using get()

print(students.get("Name"))
print(students.get("grade"))
```

```python
## If we are accessing a key that is not present in the dictionary

print(students.get("last_name"))

## If we want to print something if the key is not present in the dictionary

print(students.get("lst_name", "Not Available"))
print(students.get("Name", "Not Available"))
```

### Modifying dictionary elements

```python
students = {"Name": "cash", "age": 25, "grade": 'A'}

## Updating a value

students["grade"] = "A+"
print(students)

## Adding a key-value pair

students["country"] = "India"
print(students)

## Deleting a key-value pair

del students["age"]
print(students)
```

### Dictionary Methods

```python
## To get all the keys from dictionary
all_keys = students.keys()
print(all_keys)

## To get all the values from dictionary
all_values = students.values()
print(all_values)

## To get all the key-value pairs from dictionary
all_items = students.items()
print(all_items)
```

### Shallow Copy

```python
## Normal copy
student_copy = students
print("Before any changes")
print(students)
print(student_copy)

## After changes
students["Name"] = "cash1"
print("After changes in original dictionary")
print(students)
print(student_copy)
```

```python
## Now creating Shallow copy
print("Before any changes")
student_copy1 = students.copy()
print("Original dictionary: ", students)
print("Copy: ", student_copy1)
print()

## After changes
students["Name"] = "cash3"
print("After changes in original dictionary")
print("Original dictionary: ", students)
print("Copy: ", student_copy1)
```

### Iterating over dictionaries

```python
## Iterating over keys 

for key in students.keys():
    print(key)
```

```python
## Iterating over values

for value in students.values():
    print(value)
```

```python
## Itearting ove key-value pairs

for key, value in students.items():
    print(f"{key}: {value}")
```

### Nested Dictionaries

```python
students = {
    "student1": {"name": "cash", "age": 25},
    "student2": {"name": "krish", "age": 26}
}
print(students)
```

### Accessing nested dictionary elements

```python
print("Data of first student: ", students["student1"])
print(f"Age of first student: {students["student1"]["age"]}")
print("Name of second student: ", students["student2"]["name"])
```

### Iterating over nested dictionaries

```python
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
```

```python
for student_id, student_info in students.items():
    print(f"{student_id} name: {student_info["name"]}")
```

```python
for student_id, student_info in students.items():
    print(f"{student_id} age: {student_info["age"]}")
```

```python
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
    print()
```

### Dictionary Comprehension

```python
## Using for loop

squares = {}
for num in range(1, 11):
    squares[num] = num**2

print(squares)
```

```python
## Using dictionary comprehension

squares = {num: num**2 for num in range(1,10)}
print(squares)
```

```python
## square of even numbers

## Normally

even_sqr = {}
for x in range(1,9):
    if x%2==0:
        even_sqr[x] = x**2

print(even_sqr)
```

```python
## Using dictionary comprehension

even_sqr = {x: x**2 for x in range(1, 11) if x%2==0}
print(even_sqr)
```

### Count the frequency of elements of list using dictionary

```python
lst = [1, 2, 3, 3, 4, 4, 5, 5, 5, 6, 7, 7, 8, 8, 8, 8]
element_frequency = {x: lst.count(x) for x in lst}
print(element_frequency)
```

### Merging 2 dictionaries into 1

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```
