### Function

```python
## Defining a function
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print(f"{num} is an even number")
    else:
        print(f"{num} is an odd number")

## Calling the function
num = int(input("Enter a number: "))
even_or_odd(num)
```

```python
## Function with multiple parameters

def add(a, b):
    return a+b

result = add(2, 4)
print(result)
```

```python
## Default Parameter

def greet(name = "Person"):
    print(f"Hello {name}, how are you?")

greet("cash")
greet()
```

```python
## Variable Length Argument and Positonal argument

def print_numbers(*args): ##Variable Length Argument
    for number in args:
        print(number)
    print(args)


print_numbers(1, 2, 3, 4, 5, 6, 7, 8) ##Positonal Arguments
```

```python
## Keyword Argument

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name = "cash", age = 25, country = "India")
```

```python
## Combining both positional and keyword arguments

def print_details(*args, **kwargs):
    for val in args:
        print(f"Positional argument: {val}")

    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_details(1, 2, 3, 4, 5, 6, 7, name = "cash", age = 25, country = "India")
```

```python
## Return statement

def mult(a, b):
    return a*b

mult(2, 3)
```

```python
## We can also return multiple values

def multiplication(a, b):
    return a*b, a

print(multiplication(2, 3))
```
